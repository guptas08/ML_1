```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
```

```python
df1=pd.read_csv("D:\Machine Learning\GermanCredit\German-Credit_1.csv")
df2=pd.read_csv("D:\Machine Learning\GermanCredit\German-Credit_2.csv")
df=pd.merge(df1, df2, on=['OBS', 'OBS'])
df.head()
```

```python
df.columns
```

```python
#Count number of nulls
null_columns=df.columns[df.isnull().any()]
df[null_columns].isnull().sum()
```

```python
#All Null rows and columns
df[df.isnull().any(axis=1)][null_columns]
```

```python
#Single Column which are NULL
df[df["DURATION"].isnull()][null_columns]
df[df["AMOUNT"].isnull()][null_columns]
#df[df["SAV_ACCT"].isnull()][null_columns]

```

```python
#Filling NaN with Mean.

df['AMOUNT']=df['AMOUNT'].fillna(df['AMOUNT'].dropna().mean())
```

```python

df['AGE']=df['AGE'].fillna(df['AGE'].dropna().mean())
```

```python
#Filling NaN with 0
df['OTHER_INSTALL'].fillna(0.0,inplace=True)
df.fillna(0.0,inplace=True)
```

```python
# Ok we have to create for each group risky and non-risky loans.
df['Age_Group'] = np.nan

lst = [df]

for col in lst:
    col.loc[(col['AGE'] > 18) & (col['AGE'] <= 29), 'Age_Group'] = 'Young'
    col.loc[(col['AGE'] > 29) & (col['AGE'] <= 40), 'Age_Group'] = 'Young Adults'
    col.loc[(col['AGE'] > 40) & (col['AGE'] <= 55), 'Age_Group'] = 'Senior'
    col.loc[col['AGE'] > 55, 'Age_Group'] = 'Elder' 
    
df.head()
```
