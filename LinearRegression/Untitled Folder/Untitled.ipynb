```python
#df=pd.read_csv('C:\\Anaconda2\\envs\\py36\\lib\\site-packages\\sklearn\\datasets\\data\\boston_house_prices.csv')
df = pd.DataFrame(boston.data)
```

```python
df
```

```python
df.columns=boston.feature_names
```

```python
df.columns
```

```python
df['MEDV']=boston.target
```

```python
df
```

```python
X=df[['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX',
       'PTRATIO', 'B', 'LSTAT']].copy()
```

```python
y=df['MEDV'].copy()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
z=lr.fit(X,y)
```

```python
for col,m in zip(X.columns,lr.coef_):
    print(col,"*",m,"+")
print(lr.intercept_)
```

```python
df.describe()
```

```python
import statsmodels.api as sm

mod = sm.OLS(y,X)

fii = mod.fit()

p_values = fii.summary2().tables[1]['P>|t|']
```

```python
print((p_values.round(5)))
```

```python
p_val=pd.DataFrame()
p_val["p_values"]=p_values
p_val["features"]=X.columns
p_val["p_values"]
p_val["features"]
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif["features"] = X.columns
vif
```

```python
#df1=pd.concat([p_val,vif])
df1= pd.merge(p_val, vif, on=['features'])
```

```python

df1 = df1.reindex(columns=['features','p_values','VIF Factor'])
df1
```
