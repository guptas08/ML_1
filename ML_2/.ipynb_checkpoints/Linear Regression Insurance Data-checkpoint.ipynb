```python
def custom_func(sex,smoker):
    if sex=="male" and smoker=="yes":
        return "male_smoker"
    elif sex=="female" and smoker=="yes":
        return "female_smoker"
    elif sex=="female" and smoker=="no":
        return "female_nonsmoker"
    else:
        return "male_nonsmoker"
```

```python
df["sex_smoking"]=df[['sex','smoker']].apply(lambda x:custom_func(x['sex'],x['smoker']),axis=1)
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x="age",y="charges",hue="sex_smoking",size="bmi",data=df)
```

# Data Preparation

```python
df.columns
```

```python
X=df[['age', 'sex', 'bmi', 'children', 'smoker', 'region']].copy()
y=df['charges'].copy()
```

```python
X
```

```python
#Dummy Coding for Region where region columns in segregated into 4
```

```python
#clean the sex column
```

```python
X.sex=X.sex.map(lambda x: 1 if x=='male' else 0)
```

```python
#clean smoker column
```

```python
X.smoker=X.smoker.map(lambda x: 1 if x=='yes' else 0)
```

```python
#clean region
```

```python
dummies=pd.get_dummies(X.region)
dummies
```

```python
X
```

```python
X=X.join(dummies)
X=X.drop(["region"],axis=1)
```

```python
X
```

# Model

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X,y)
```

```python
for col,m in zip(X.columns,lr.coef_):
    print(col,"*",m,"+")
print(lr.intercept_)
```

```python
y_hat=lr.predict(X)
residual=(y-y_hat)
```

```python
sum(residual**2)/len(y)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
print("MSE",mean_squared_error(y,y_hat))
```

```python
print("R2 Error",r2_score(y,y_hat))
```

# Testing the Regression Model

```python
#Assumption 1 : Sum of Residual should be Zero
sum(residual)
```

```python
#Assumption 2 : y-y_hat should follow normal dist where mean=0
sns.distplot(residual)
```

```python
#Assumption 3 : look for homoscadacity. points should be around the lines
plt.scatter(y_hat,residual)
plt.hlines(y=0,xmin=-1000,xmax=50000)
```

```python
#Assumption 4 : Real vs the predicted value

sns.scatterplot(y,y_hat)
```

```python
df[abs(residual)>10000]
```

```python
def fit_smoking(smoker,bmi):
    if smoker=="yes" and bmi >30:
        return "Obese_Smoker"
    elif smoker=="yes" and bmi <=30:
        return "fit_smoker"
    elif smoker =="no" and bmi>30:
        return "fat"
    else:
        return "fit"
    
    
```

```python
df["Smoking_Fit"]=df[["smoker","bmi"]].apply(lambda x: fit_smoking(x["smoker"],x["bmi"]),axis=1)
```

```python
sns.scatterplot(x="age",y="charges",hue="Smoking_Fit",data=df)
```

```python
sns.scatterplot(x="bmi",y="charges",hue="Smoking_Fit",data=df)
```

# Model 2

```python
df.columns
```

```python
X=df[['age', 'sex', 'bmi', 'children', 'region', 'Smoking_Fit']].copy()
y=df['charges'].copy()
```

```python
X.sex=X.sex.map(lambda x: 1 if x=="male" else 0)
X=X.join(pd.get_dummies(X.region)).drop(['region'],axis=1)
X=X.join(pd.get_dummies(X.Smoking_Fit)).drop(['Smoking_Fit'],axis=1)
```

```python
lr=LinearRegression()
lr.fit(X,y)
y_hat=lr.predict(X)
residual=y-y_hat
print("R-Square",r2_score(y,y_hat))
```

```python
for m,col in zip(lr.coef_,X.columns):
    print(m,"*",col,"+")
print(lr.intercept_)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
#Assumption 2 : y-y_hat should follow normal dist where mean=0

sns.distplot(residual)
```

```python
#Assumption 3 : look for homoscadacity. points should be around the lines
plt.scatter(y_hat,residual)
plt.hlines(y=0,xmin=-1000,xmax=50000)
```

```python
#Assumption 4 : Real vs the predicted value

sns.scatterplot(y,y_hat)
```

# Cross Validation

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.2,random_state=2)
```

```python
X
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
lr=LinearRegression()
lr.fit(X_train,y_train)
print("R-Squared Error Training",lr.score(X_train,y_train))
print("R-Squared Error Testing",lr.score(X_test,y_test))
```

```python
plt.scatter(y_train,lr.predict(X_train),color="r")
plt.scatter(y_test,lr.predict(X_test),color="g")
```

```python
for m,col in zip(lr.coef_,X_train.columns):
    print(m,"*",col,"+")
print(lr.intercept_)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=2, include_bias=False)
X_poly_train=poly.fit_transform(X_train)
X_poly_test=poly.transform(X_test)
lr=LinearRegression()
lr.fit(X_poly_train,y_train)
print("R_squared Error for training",lr.score(X_poly_train,y_train))
print("R_squared Error for testing",lr.score(X_poly_test,y_test))
```
