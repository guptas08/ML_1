```python
X=np.arange(20)
Y=4*X+20
```

```python
plt.scatter(X,Y)
plt.xlabel("X")
plt.ylabel("Y")
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X.reshape(-1,1),Y)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
y_hat=lr.predict(X.reshape(-1,1))
```

```python
plt.scatter(X,Y)
plt.plot(y_hat)
plt.xlabel("X")
plt.ylabel("Y")
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
print("MSE",mean_squared_error(Y,y_hat))
```

```python
print("R-Square Error",r2_score(Y,y_hat))
```
