```python
import matplotlib.pyplot as plt
```

```python


x=[1,2,3,4,5,6]
y=[8,7,8,5,9,5]

x1=[1,3,5,7,4]
y1=[5,8,7,2,6]
```

```python
plt.plot(x,y,label='Line1')
plt.plot(x1,y1,label='Line2')


plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend()
```

```python
plt.bar(x,y)
plt.bar(x1,y1)
```

```python
plt.scatter(x,y,label='Add',marker='+',color='k',s=100)
plt.scatter(x1,y1,marker='*',label='Star',color='r',s=55)
```
